```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('CreditCardData.csv')
print("testing")
```

```python
encoder  = LabelEncoder()
```

```python
import os

df['Attrition_Flag'] = encoder.fit_transform(df['Attrition_Flag'])
df["Education_Level"] = encoder.fit_transform(df['Education_Level'])
df["Gender"] = encoder.fit_transform(df["Gender"])
df["Marital_Status"] = encoder.fit_transform(df["Marital_Status"])
df['Income_Category'] = encoder.fit_transform(df['Income_Category'])
df['Card_Category'] = encoder.fit_transform(df['Card_Category'])

filename = 'Encoded_CreditCard_Data.csv'

if not os.path.exists(filename):
    df.to_csv(filename, index=False)
    print(f"DataFrame saved as {filename}")
else:
    print(f"{filename} already exists. DataFrame not saved.")
```

```python
df.info()
```

```python
features = ['Attrition_Flag', 'Customer_Age', 'Gender',
       'Dependent_count', 'Education_Level', 'Marital_Status',
       'Income_Category', 'Card_Category', 'Months_on_book',
       'Total_Relationship_Count', 'Months_Inactive_12_mon',
       'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
       'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1',
       'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio']
target = ['Total_Trans_Amt']
```

```python
X= df[features]
y=df[target]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn import metrics

def train_test(X_train, X_test, y_train, y_test, param_grid, reg):
    reg.fit(X_train, y_train)

    print("*** Parameter estimation results: ")
    print(reg.cv_results_)
    print()

    print("*** Grid scores: ")
    means = reg.cv_results_['mean_test_score']
    stds = reg.cv_results_['std_test_score']
    params = reg.cv_results_['params']

    for mean, std, param in zip(means, stds, params):
        print(f"{round(mean, 3)} (+/-{round(std*2, 3)}) for {param}")
    print()

    print("*** Highest R^2 score: ")
    print(f"{round(means.max(), 3)}")
    print()

    print("*** Best parameters set found: ")
    print(reg.best_params_)
    print()

    print("*** Final R^2 score: ")
    y_true, y_pred = y_test, reg.predict(X_test)
    r2_score = round(metrics.r2_score(y_true, y_pred), 3)
    print(r2_score)

    return reg, r2_score
```

```python
cv = 5
summary = dict()
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression

param_grid = [{"fit_intercept": [True, False]}]
lr = GridSearchCV(LinearRegression(), param_grid, cv=cv)
lr, score = train_test(X_train_scaled, X_test_scaled, y_train, y_test, param_grid, lr)
```
