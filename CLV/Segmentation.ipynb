```python
import pandas as pd
```

```python
df = pd.read_csv('Encoded_CreditCard_Data.csv')
```

```python
exsisting = df[df['Attrition_Flag'] == 1]
```

```python
exsisting.drop('CLIENTNUM', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Assuming df is your DataFrame and you want to normalize column 'A'
scaler = MinMaxScaler()

# Note that MinMaxScaler expects a 2D array so we use df[['A']] to get a DataFrame, not a Series
exsisting['Credit_Limit_normalized'] = scaler.fit_transform(exsisting[['Credit_Limit']])
exsisting['Avg_Utilization_Ratio_normalized'] = scaler.fit_transform(exsisting[['Avg_Utilization_Ratio']])

exsisting.head()
```

```python
exsisting['Credit_Score'] = exsisting['Credit_Limit_normalized'] * exsisting['Avg_Utilization_Ratio_normalized']
```

```python
exsisting['Credit_Group'] = pd.cut(exsisting['Credit_Score'], bins=[0, 0.2, 0.4, 0.6, 0.8, 1], labels=[0,1,2,3,4])
labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
credit_score_key = {labels[i]: i for i in range(len(labels))}
```
