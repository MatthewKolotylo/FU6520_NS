```python
# Load the data
df = pd.read_csv('CreditCardData.csv')

# change the data type of the columns to float
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Existing Customer': 0, 'Attrited Customer': 1})
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0})
df['Education_Level'] = df['Education_Level'].map({'High School': 0, 'Graduate': 1, 'Uneducated': 2, 'College': 3, 'Post-Graduate': 4, 'Doctorate': 5, 'Unknown': 6})
df['Marital_Status'] = df['Marital_Status'].map({'Married': 0, 'Single': 1, 'Divorced': 2, 'Unknown': 3})
df['Income_Category'] = df['Income_Category'].map({'Less than $40K': 0, '$40K - $60K': 1, '$60K - $80K': 2, '$80K - $120K': 3, '$120K +': 4, 'Unknown': 5})
df['Card_Category'] = df['Card_Category'].map({'Blue': 0, 'Silver': 1, 'Gold': 2, 'Platinum': 3})
```

```python
non_zero_count = df[df['Total_Revolving_Bal'] != 0].shape[0]
total_count = df.shape[0]
percentage_all_customers = (non_zero_count / total_count) * 100


# Calculate the percentage of non-zero revolving balance for attrited customers
attrited_non_zero_count = Attrited_Customers[Attrited_Customers['Total_Revolving_Bal'] != 0].shape[0]
attrited_percentage = (attrited_non_zero_count / Attrited_Customers.shape[0]) * 100

# Calculate the percentage of non-zero revolving balance for existing customers
existing_non_zero_count = Existing_Customers[Existing_Customers['Total_Revolving_Bal'] != 0].shape[0]
existing_percentage = (existing_non_zero_count / Existing_Customers.shape[0]) * 100

# Create a dictionary to store the percentages
percentages = {
    'All Customers': percentage_all_customers,
    'Attrited Customers': attrited_percentage,
    'Existing Customers': existing_percentage
}

percentages
```

### Logistic Regression

```python
df_clean = Existing_Customers.append(non_zero_balance_attrited_customers)
df_clean.shape
```

```python
from sklearn.preprocessing import LabelEncoder
encoder  = LabelEncoder()

df_clean['Attrition_Flag'] = encoder.fit_transform(df_clean['Attrition_Flag'])
df_clean["Education_Level"] = encoder.fit_transform(df_clean['Education_Level'])
df_clean["Gender"] = encoder.fit_transform(df_clean["Gender"])
df_clean["Marital_Status"] = encoder.fit_transform(df_clean["Marital_Status"])
df_clean['Income_Category'] = encoder.fit_transform(df_clean['Income_Category'])
df_clean['Card_Category'] = encoder.fit_transform(df_clean['Card_Category'])

df_clean = df_clean.drop('CLIENTNUM', axis=1)
```

```python
df_clean
```

```python
from sklearn.model_selection import train_test_split

df_clean_X = df_clean[df_clean.columns[1:]]
df_clean_y = df_clean[df_clean.columns[0]]

X_train, X_test, y_train, y_test = train_test_split(df_clean_X, df_clean_y , test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)

RFC.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred = pd.DataFrame(RFC.predict(X_test))
y_pred_proba = pd.DataFrame(RFC.predict_proba(X_test)[:, 1])
#y_pred_proba = y_pred_proba.rename(columns={0: 'Attrited Customer'})
#y_pred_proba = y_pred_proba.rename(columns={1: 'Existing Customer'})
```

```python
y_pred.value_counts()
```

```python
y_pred_proba
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Print performance metrics
print(f'Accuracy: {accuracy}')
print(f'ROC-AUC: {roc_auc}')
```

```python
year_1_demo = ['Customer_Age', 'Gender',
       'Dependent_count', 'Education_Level', 'Marital_Status',
       'Income_Category', 'Card_Category', 'Months_on_book',
       'Total_Relationship_Count', 'Months_Inactive_12_mon',
       'Contacts_Count_12_mon']
predict = ['Credit_Limit']
```

```python
year_1_X = year_1[year_1_demo]
year_1_Y = year_1[predict]
```

```python
year_2_X = year_1_X.copy()
year_2_X['Customer_Age'] = year_2_X['Customer_Age'] + 1
year_2_X['Months_on_book'] = year_2_X['Months_on_book'] + 12
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(year_1_X, year_1_Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor

# Create the classifier
RFR = RandomForestRegressor()

# Fit the classifier to the training data
RFR.fit(X_train, y_train)

# Predict on the test data
year_1_credit_limit_pred = RFR.predict(X_test)
```

```python
score = RFR.score(X_test, y_test)
print("Model Score:", score)
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

RFR = RandomForestRegressor()

# Create the GridSearchCV object
grid_search = GridSearchCV(estimator=RFR, param_grid=param_grid, cv=5, n_jobs=-1)

# Fit the GridSearchCV object to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

# Print the best parameters and best score
print("Best Parameters:", best_params)
print("Best Score:", best_score)
```
